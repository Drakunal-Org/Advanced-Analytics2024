```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import json
import regex as re
```

```python
response =  requests.get('https://www.yelp.com/search?find_desc=restaurants&find_loc=San+Francisco%2C+CA')
```

```python
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
allPrices = soup.findAll("span",{"class": "css-chan6m"})
```

```python
soup.get_text()
```

```python
print(allPrices)
```

```python
for price in allPrices:
    print(price.get_text())
```

```python
items = soup.findAll("a", {"class": "title"})
```

```python
print(items)
```

```python
count = 0
for item in items:
    print(item.get_text())
    count = count + 1
```

```python
list_prices = []
for price in allPrices:
    list_prices.append(price.get_text())
```

```python
print(list_prices)
```

```python
list_items = []
for item in items:
    list_items.append(item.get_text())
```

```python
print(list_items)
```

```python
df = pd.DataFrame(
    {'Product': list_items,
     'Price': list_prices
    })
```

```python
print(df)
```

```python
json_data = df.to_json()
print(json_data)
```

```python
with open("sample.json", "w") as outfile:
    outfile.write(json_data)
```

```python
zip_iterator = zip(list_items, list_prices) # gets the pair of the elements
```

```python
dictionary_x = dict(zip_iterator) #converts it to a dictionary
```

```python
print(dictionary_x)
```

```python
json_object = json.dumps(dictionary_x, indent = 4) 
print(json_object)
```

```python
with open("sample2.json", "w") as outfile:
    outfile.write(json_object)
```

```python
for item in list_items:
    print(item)
```

```python
print(list_items)
```

```python
better_name = []
for item in list_items:
    item = re.sub('\.', '', item)
    print(item)
    better_name.append(item)
```

```python
print(better_name)
```

```python
better_price = []
for price in list_prices:
    price = re.sub('\$', '', price)
    print(price)
    better_price.append(float(price))
```

```python
zip_iterator2 = zip(better_name,better_price) # gets the pair of the elements
```

```python
dictionary_y = dict(zip_iterator2) #converts it to a dictionary
```

```python
json_object2 = json.dumps(dictionary_y, indent = 4) 
print(json_object2)
```

```python
with open("sample3.json", "w") as outfile:
    outfile.write(json_object2)
```
